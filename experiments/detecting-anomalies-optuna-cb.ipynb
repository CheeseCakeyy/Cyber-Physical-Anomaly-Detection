## Importing data:

```python
import gc
import polars as pl

train_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/train.csv'
test_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/test.csv'

train_df = pl.read_csv(train_path)
print('Training set: ', train_df.shape)
display(train_df.head(3))
```

```python
#Finding number of unique values in each column
unique_dict = {col:train_df[col].n_unique() for col in train_df.columns}

#removing redundant columns
unique_count_1_cols = [col for col,count in unique_dict.items() if count==1]
train_df = train_df.drop(unique_count_1_cols)
print(f"Dropped a total of {len(unique_count_1_cols)} columns \n")
print("Dropped columns:", unique_count_1_cols)
```

```python
#Dropping columns with too many null values
cols_to_drop = ['DERCtlAC[0].PFWAbs.Ext','DERCtlAC[0].PFWAbsRvrt.Ext','DERMeasureAC[0].ThrotPct','DERMeasureAC[0].ThrotSrc']
train_df = train_df.drop(cols_to_drop)
```

```python
#Dropping null values from remaining columns
train_df = train_df.drop_nulls()
```

```python
#Seprating label and features
X = train_df.drop(['Id','Label'])
y = train_df['Label']
```

```python
#clearing up memory since the dataset is huge
del train_df
gc.collect()
print("Memory cleared!")
```

```python
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val = train_test_split(X,y,
                                               test_size=0.2,
                                               stratify=y,
                                               random_state=42)
```

```python
cat_cols = X.select(pl.col(pl.Utf8)).columns
```

```python
#clearing up memory since the dataset is huge
del X,y
gc.collect()
print("Memory cleared!")
```

## Preprocessing:

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

cat_pipe = Pipeline([
('encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer(
    transformers=[('cat', cat_pipe, cat_cols)],
    remainder='passthrough'
)
```

```python
X_processed = preprocessor.fit_transform(X_train)
X_val_processed = preprocessor.transform(X_val)
```

```python
#CatBoost doesnt support polars series so...
y_train_np = y_train.to_numpy()
y_val_np = y_val.to_numpy()
```

```python
del X_train, X_val
gc.collect()
print('Memory Cleared!')
```

## Optuna Tuning:

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning) #warnings are annoying so this bypasses them
```

```python
import optuna
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import fbeta_score

def objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 500, 3500),
        'depth': trial.suggest_int('depth', 4, 8),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 50, log=True),
    
        'task_type': 'GPU',
        'devices': '0',
        'verbose': False,
        'random_seed': 42,
    }
    
    model = CatBoostClassifier(**params)
   
    model.fit(X_processed, y_train_np)
    
    y_probs = model.predict_proba(X_val_processed)[:, 1]
    
    best_score = 0
    for t in [i/100 for i in range(10, 90)]:
        preds = (y_probs > t).astype(int)
        score = fbeta_score(y_val_np, preds, beta=2)
        
        if score > best_score:
            best_score = score

    return best_score
```

## Tuned CB:

```python
tuned_params =  {'iterations': 2525,
           'depth': 8,
           'learning_rate': 0.06706560941131733,
           'l2_leaf_reg': 18.231582992471346}

default_params = {'task_type': 'GPU',
        'devices': '0',
        'verbose': False,
        'random_seed': 42,}
```

```python
model = CatBoostClassifier(
    **tuned_params,
    **default_params
)
```

```python
model.fit(X_processed, y_train_np)
```

```python
#finding the best threshold
y_val_perd_probs = model.predict_proba(X_val_processed)[:, 1]

best_score = 0
best_thresh = 0

for t in [i/100 for i in range(10, 90)]:
    preds = (y_val_perd_probs > t).astype(int)
    score = fbeta_score(y_val_np, preds, beta=2) #bets=2 stands for f2_score which is our metric for this competition
    
    if score > best_score:
        best_score = score
        best_thresh = t

print(best_thresh, best_score)
```

## Making predictions on test data:

```python
test_df = pl.read_csv(test_path)
print('Test set: ', test_df.shape)
display(test_df.head(3))
```

```python
cols_to_remove = unique_count_1_cols + cols_to_drop + ['Id']
X_test = test_df.drop(cols_to_remove)
X_test_processsed = preprocessor.transform(X_test)
```

```python
y_test_proba = model.predict_proba(X_test_processsed)[:,1]

y_preds = (y_test_proba > best_thresh).astype(int)
```

```python
submission = pl.DataFrame({
    "Id": test_df['Id'],
    "Label": y_preds
})

submission.write_csv("submission_CB.csv")
```
