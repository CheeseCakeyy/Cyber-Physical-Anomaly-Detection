## Importing data:

```python
import gc
import polars as pl

train_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/train.csv'
test_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/test.csv'

train_df = pl.read_csv(train_path)
print('Training set: ', train_df.shape)
display(train_df.head(3))
```

```python
#Finding number of unique values in each column
unique_dict = {col:train_df[col].n_unique() for col in train_df.columns}

#removing redundant columns
unique_count_1_cols = [col for col,count in unique_dict.items() if count==1]
train_df = train_df.drop(unique_count_1_cols)
print(f"Dropped a total of {len(unique_count_1_cols)} columns \n")
print("Dropped columns:", unique_count_1_cols)
```

```python
#Dropping columns with too many null values
cols_to_drop = ['DERCtlAC[0].PFWAbs.Ext','DERCtlAC[0].PFWAbsRvrt.Ext','DERMeasureAC[0].ThrotPct','DERMeasureAC[0].ThrotSrc']
train_df = train_df.drop(cols_to_drop)
```

```python
#Dropping null values from remaining columns
train_df = train_df.drop_nulls()
```

```python
cat_cols = list(train_df.select(pl.col(pl.Utf8)).columns)
```

```python
#Seprating label and features
X = train_df.drop(['Id','Label'])
y = train_df['Label']
```

```python
#clearing up memory since the dataset is huge
del train_df
gc.collect()
print("Memory cleared!")
```

```python
import polars.selectors as cs

def digit_and_mod_features(df, cols, max_digits=4):
    new_cols = []
    
    for col in cols:
        for i in range(max_digits):
            new_cols.append(((pl.col(col) // (10**i)) % 10).alias(f"{col}_digit_{i}"))
        
        # Mod features (very powerful)
        new_cols.append((pl.col(col) % 10).alias(f"{col}_mod10"))
        new_cols.append((pl.col(col) % 50).alias(f"{col}_mod50"))
        new_cols.append((pl.col(col) % 100).alias(f"{col}_mod100"))

    return df.with_columns(new_cols)


# Select numeric columns
num_cols = X.select(cs.numeric()).columns

# Use more columns (try 20 instead of 12)
digit_cols = num_cols[:12]

# Apply
X = digit_and_mod_features(X, digit_cols, max_digits=4)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val = train_test_split(X,y,
                                               test_size=0.2,
                                               stratify=y,
                                               random_state=42)
```

```python
#clearing up memory since the dataset is huge
del X,y
gc.collect()
print("Memory cleared!")
```

## Preprocessing:

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

cat_pipe = Pipeline([
('encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer(
    transformers=[('cat', cat_pipe, cat_cols)],
    remainder='passthrough'
)
```

```python
X_processed = preprocessor.fit_transform(X_train)
X_val_processed = preprocessor.transform(X_val)
```

```python
del X_train, X_val
gc.collect()
print('Memory Cleared!')
```

## Optuna Tuning:

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning) #warnings are annoying so this bypasses them
```

```python
import optuna
from lightgbm import LGBMClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import fbeta_score

def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 2500, 4500),
        "learning_rate": trial.suggest_float("learning_rate", 0.005, 0.2, log=True),
        
        "num_leaves": trial.suggest_int("num_leaves", 16, 128),
        "max_depth": trial.suggest_int("max_depth", 3, 12),
        
        "min_child_samples": trial.suggest_int("min_child_samples", 10, 100),
        "subsample": trial.suggest_float("subsample", 0.6, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.6, 1.0),
        
        "reg_alpha": trial.suggest_float("reg_alpha", 1e-3, 10, log=True),
        "reg_lambda": trial.suggest_float("reg_lambda", 1e-3, 10, log=True),

        "random_state": 42,
        "n_jobs": -1,
        "verbose":-1
    }
    
    model = LGBMClassifier(**params)
    
    model.fit(X_processed, y_train)
    
    y_probs = model.predict_proba(X_val_processed)[:, 1]
    
    best_score = 0
    for t in [i/100 for i in range(10, 90)]:
        preds = (y_probs > t).astype(int)
        score = fbeta_score(y_val, preds, beta=2)
        
        if score > best_score:
            best_score = score

    return best_score
```

```python
tuned_lgbm_params ={'n_estimators': 3774,
                    'learning_rate': 0.043701760061425884,
                    'num_leaves': 51,
                    'max_depth': 8,
                    'min_child_samples': 60, 
                    'subsample': 0.9375313270803008, 
                    'colsample_bytree': 0.731959413738176,
                    'reg_alpha': 3.1921940864316802,
                    'reg_lambda': 0.02530533768874328} 


default_params = {
        "random_state": 42,
        "n_jobs": -1,
        "verbose":-1
}
```

```python
from lightgbm import LGBMClassifier

model = LGBMClassifier(
    **tuned_lgbm_params,
    **default_params
)
```

```python
model.fit(X_processed, y_train)
```

```python
from sklearn.metrics import fbeta_score

#finding the best threshold
y_val_perd_probs = model.predict_proba(X_val_processed)[:, 1]

best_score = 0
best_thresh = 0

for t in [i/100 for i in range(10, 90)]:
    preds = (y_val_perd_probs > t).astype(int)
    score = fbeta_score(y_val, preds, beta=2) #bets=2 stands for f2_score which is our metric for this competition
    
    if score > best_score:
        best_score = score
        best_thresh = t

print(best_thresh, best_score)
```
