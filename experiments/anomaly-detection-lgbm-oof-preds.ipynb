## Importing data:

```python
import gc
import polars as pl

train_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/train.csv'
test_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/test.csv'

train_df = pl.read_csv(train_path)
print('Training set: ', train_df.shape)
display(train_df.head(3))
```

```python
#Finding number of unique values in each column
unique_dict = {col:train_df[col].n_unique() for col in train_df.columns}

#removing redundant columns
unique_count_1_cols = [col for col,count in unique_dict.items() if count==1]
train_df = train_df.drop(unique_count_1_cols)
print(f"Dropped a total of {len(unique_count_1_cols)} columns \n")
print("Dropped columns:", unique_count_1_cols)
```

```python
#Dropping columns with too many null values
cols_to_drop = ['DERCtlAC[0].PFWAbs.Ext','DERCtlAC[0].PFWAbsRvrt.Ext','DERMeasureAC[0].ThrotPct','DERMeasureAC[0].ThrotSrc']
train_df = train_df.drop(cols_to_drop)
```

```python
#Dropping null values from remaining columns
train_df = train_df.drop_nulls()

#Seprating label and features
X = train_df.drop(['Id','Label'])
y = train_df['Label']

#clearing up memory since the dataset is huge
del train_df
gc.collect()
print("Memory cleared!")
```

## Preprocessing:

```python
cat_cols = X.select(pl.col(pl.Utf8)).columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

cat_pipe = Pipeline([
('encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer(
    transformers=[('cat', cat_pipe, cat_cols)],
    remainder='passthrough'
)

X_processed = preprocessor.fit_transform(X)
print('Preprocessing completed')
```

```python
del X
gc.collect()
print('Memory Cleared!')
```

```python
#importing test data:
test_df = pl.read_csv(test_path)
print('Test set: ', test_df.shape)
display(test_df.head(3))
```

```python
cols_to_remove = unique_count_1_cols + cols_to_drop + ['Id']
X_test = test_df.drop(cols_to_remove)
X_test_processed = preprocessor.transform(X_test)
```

## Best HyperParams:
These were found out using optuna

```python
tuned_lgbm_params = {'n_estimators': 3401,
                     'learning_rate': 0.020009612153555324, 
                     'num_leaves': 76, 
                     'max_depth': 10, 
                     'min_child_samples': 35,
                     'subsample': 0.9974353945552556,
                     'colsample_bytree': 0.6161939987092331,
                     'reg_alpha': 0.1695611207180594,
                     'reg_lambda': 0.16021326556236573}

default_params = {
        "random_state": 42,
        "n_jobs": -1,
        "verbose":-1
}
```

## OOF predictions:

```python
X_np = X_processed
y_np = y
X_test_np = X_test_processed

del X_processed, y, X_test_processed, X_test, test_df
gc.collect()
print('Memory cleared!')
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import fbeta_score
import numpy as np
from lightgbm import LGBMClassifier

oof_preds = np.zeros(len(X_np))
test_preds = np.zeros(len(X_test_np))

fold_scores = []
fold_thresholds = []

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for fold, (train_idx, val_idx) in enumerate(skf.split(X_np, y_np)):

    X_tr, X_val = X_np[train_idx], X_np[val_idx]
    y_tr, y_val = y_np[train_idx], y_np[val_idx]

    model = LGBMClassifier(
        **tuned_lgbm_params,
        **default_params
    )

    model.fit(X_tr, y_tr)

    val_preds = model.predict_proba(X_val)[:, 1]
    oof_preds[val_idx] = val_preds

    test_preds += model.predict_proba(X_test_np)[:, 1] / 5

    # Threshold tuning per fold
    best_thr = 0
    best_score = 0

    for thr in np.linspace(0.01, 0.5, 100):
        score = fbeta_score(y_val, (val_preds > thr).astype(int), beta=2)
        if score > best_score:
            best_score = score
            best_thr = thr

    fold_scores.append(best_score)
    fold_thresholds.append(best_thr)

    print(f"Fold {fold} | F2: {best_score:.5f} | Thr: {best_thr:.4f}")

print("Mean Fold F2:", np.mean(fold_scores))
print("Fold Thresholds:", fold_thresholds)
```

```python
del X_np, X_test_np
gc.collect()
print('Memory cleared!')
```

```python
#Global best threshold
best_thr = 0
best_score = 0

thresholds = np.percentile(oof_preds, np.linspace(1, 99, 200))

for thr in thresholds:
    score = fbeta_score(y_np, (oof_preds > thr).astype(int), beta=2)
    if score > best_score:
        best_score = score
        best_thr = thr

print("Global Threshold:", best_thr)
print("OOF F2:", best_score)

#final preds
final_preds = (test_preds > best_thr).astype(int)
```

```python
test_df = pl.read_csv(test_path)
```

```python
submission = pl.DataFrame({
    'Id':test_df['Id'],
    'Label':final_preds
})

submission.write_csv('Submission_OOF_preds.csv')
```
