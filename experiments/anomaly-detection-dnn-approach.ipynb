## Importing data:

```python
import gc
import polars as pl

train_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/train.csv'

train_df = pl.read_csv(train_path)
print('Training set: ', train_df.shape)
display(train_df.head(3))
```

```python
#Finding number of unique values in each column
unique_dict = {col:train_df[col].n_unique() for col in train_df.columns}

#removing redundant columns
unique_count_1_cols = [col for col,count in unique_dict.items() if count==1]
train_df = train_df.drop(unique_count_1_cols)
print(f"Dropped a total of {len(unique_count_1_cols)} columns \n")
print("Dropped columns:", unique_count_1_cols)
```

```python
#Dropping columns with too many null values
cols_to_drop = ['DERCtlAC[0].PFWAbs.Ext','DERCtlAC[0].PFWAbsRvrt.Ext','DERMeasureAC[0].ThrotPct','DERMeasureAC[0].ThrotSrc']
train_df = train_df.drop(cols_to_drop)
```

```python
#Dropping null values from remaining columns
train_df = train_df.drop_nulls()
```

```python
#Seprating label and features
X = train_df.drop(['Id','Label'])
y = train_df['Label']
```

```python
#clearing up memory since the dataset is huge
del train_df
gc.collect()
print("Memory cleared!")
```

## Feature Engineering:

### Domain Specific Feature Engineering:
1. Power & Current Interaction Features
   * power_factor
   * power_per_amp
2. Capacity-Based Features
   * power_utilization

```python
#Reusable feature engineering utility

def create_features(df):
    df = df.with_columns([
        #Power and current interaction features
        (pl.col("DERMeasureAC[0].W") / (pl.col("DERMeasureAC[0].VA") + 1e-6)).alias("power_factor"),
        (pl.col("DERMeasureAC[0].W") / (pl.col("DERMeasureAC[0].A") + 1e-6)).alias('power_per_amp'),

        #Capacity based features
        (pl.col("DERMeasureAC[0].W") / (pl.col("DERCapacity[0].WMax") + 1e-6)).alias("power_utilization"),
    ])

    print('3 Domain specific features created!')

    return df
```

```python
X = create_features(X)
```

## Preprocessing:

```python
import polars.selectors as cs

cat_cols = X.select(pl.col(pl.Utf8)).columns
num_cols = X.select(cs.numeric()).columns
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(handle_unknown='ignore'), cat_cols),
        ('scaler', StandardScaler(), num_cols)
    ],
    remainder='passthrough'
)
```

```python
X_processed = preprocessor.fit_transform(X)
```

```python
del X
gc.collect()
print('Memory Cleared!')
```

## Splitting data:

```python
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val = train_test_split(X_processed,y,
                                               test_size=0.2,
                                               stratify=y,
                                               random_state=42)
```

```python
print(X_train.shape) #to see what to specify as input to the model
```

```python
del X_processed, y
gc.collect()
print('Memory Cleared!')
```

## Model Architecture:

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

inputs = keras.Input(shape=(465,), name='inputs')

x = layers.BatchNormalization()(inputs)

x = layers.Dense(256,activation='relu') (x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3) (x)

x = layers.Dense(128,activation='relu') (x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3) (x)

x = layers.Dense(64, activation='relu') (x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3) (x)

outputs = layers.Dense(1, activation='sigmoid', name='outputs') (x)

model = keras.Model(inputs, outputs)
```

```python
#custom metric
def f2_score(y_true, y_pred):
    y_true = tf.cast(y_true, tf.float32)  
    y_pred = tf.cast(y_pred > 0.2, tf.float32)

    tp = tf.reduce_sum(y_true * y_pred)
    fp = tf.reduce_sum((1 - y_true) * y_pred)
    fn = tf.reduce_sum(y_true * (1 - y_pred))

    precision = tp / (tp + fp + 1e-7)
    recall = tp / (tp + fn + 1e-7)

    return (5 * precision * recall) / (4 * precision + recall + 1e-7)
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[f2_score]
)

model.summary()
```

```python
callbacks = [
    keras.callbacks.ModelCheckpoint('best_model.keras', save_best_only=True)
]

history = model.fit(
                X_train, y_train,
                validation_data=(X_val, y_val),
                epochs=25,
                callbacks=callbacks,
                batch_size=512
            )
```

```python
from sklearn.metrics import fbeta_score

model_best = keras.models.load_model("best_model.keras", 
                                     custom_objects={"f2_score": f2_score},
                                     safe_mode=False)

y_prob = model_best.predict(X_val)
y_pred = (y_prob > 0.1).astype(int)
f2 = fbeta_score(y_val, y_pred, beta=2)
print(f2)
```

```python
del X_train,y_train,X_val,y_val
gc.collect()
print('Memory Cleared!')
```

## Test data:

```python
test_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/test.csv'

test_df = pl.read_csv(test_path)
cols_to_remove = unique_count_1_cols + cols_to_drop + ['Id']
X_test = test_df.drop(cols_to_remove)
X_test = create_features(X_test)
```

```python
X_test_processed = preprocessor.transform(X_test)
```

```python
y_test_proba = model_best.predict(X_test_processed).flatten()

y_preds = (y_test_proba > 0.1).astype(int)
```

```python
submission = pl.DataFrame({
    "Id": test_df['Id'],
    "Label": y_preds
})

submission.write_csv("submission_DNN.csv")
```
