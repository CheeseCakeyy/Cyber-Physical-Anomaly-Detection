## Importing data:

```python
import gc
import polars as pl

train_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/train.csv'

train_df = pl.read_csv(train_path)
print('Training set: ', train_df.shape)
display(train_df.head(3))
```

```python
#Finding number of unique values in each column
unique_dict = {col:train_df[col].n_unique() for col in train_df.columns}

#removing redundant columns
unique_count_1_cols = [col for col,count in unique_dict.items() if count==1]
train_df = train_df.drop(unique_count_1_cols)
print(f"Dropped a total of {len(unique_count_1_cols)} columns \n")
print("Dropped columns:", unique_count_1_cols)
```

```python
#counting columns by dtypes
from collections import Counter
import matplotlib.pyplot as plt

data_types = [str(dtype) for dtype in train_df.dtypes]
counts = Counter(data_types)
print(counts)

plt.bar(counts.keys(),counts.values())
plt.xlabel('dtypes')
plt.ylabel('counts')
plt.show()
```

```python
nulls = (
    train_df.null_count()
    .transpose(include_header=True)
    .rename({"column": "feature", "column_0": "null_count"})
    .filter(pl.col("null_count") > 0)
)

pl.Config.set_tbl_rows(-1) #so that everything gets printed
print(nulls)
```

```python
#Dropping columns with too many null values
cols_to_drop = ['DERCtlAC[0].PFWAbs.Ext','DERCtlAC[0].PFWAbsRvrt.Ext','DERMeasureAC[0].ThrotPct','DERMeasureAC[0].ThrotSrc']
train_df = train_df.drop(cols_to_drop)
```

```python
#Dropping null values from remaining columns
train_df = train_df.drop_nulls()
```

```python
#Seprating label and features
X = train_df.drop(['Id','Label'])
y = train_df['Label']
```

```python
#clearing up memory since the dataset is huge
del train_df
gc.collect()
print("Memory cleared!")
```

## Domain Specific Feature Engineering:
1. Power & Current Interaction Features
   * power_factor
   * power_per_amp
2. Capacity-Based Features
   * power_utilization
   * current_utilization
3. Frequency & Harmonics
   * freq_deviation
   * phase_imbalance

```python
#Reusable feature engineering utility

def create_features(df):
    df = df.with_columns([
        #Power and current interaction features
        (pl.col("DERMeasureAC[0].W") / (pl.col("DERMeasureAC[0].VA") + 1e-6)).alias("power_factor"),
        (pl.col("DERMeasureAC[0].W") / (pl.col("DERMeasureAC[0].A") + 1e-6)).alias('power_per_amp'),

        #Capacity based features
        (pl.col("DERMeasureAC[0].W") / (pl.col("DERCapacity[0].WMax") + 1e-6)).alias("power_utilization"),
        (pl.col("DERMeasureAC[0].A") / (pl.col("DERCapacity[0].AMax") + 1e-6)).alias("current_utilization"),

        #Freq and Harmonics
        (pl.col("DERMeasureAC[0].Hz") - 50).abs().alias("freq_deviation"),  # assuming 50Hz system
        (pl.col("DERMeasureAC[0].AL1") - pl.col("DERMeasureAC[0].AL2")).abs().alias("phase_imbalance"),
    ])

    print('3 Domain specific features created!')

    return df
```

```python
X = create_features(X)
```

## Splittig data:

```python
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val = train_test_split(X,y,
                                               test_size=0.2,
                                               stratify=y,
                                               random_state=42)
```

```python
cat_cols = X.select(pl.col(pl.Utf8)).columns
```

```python
#clearing up memory since the dataset is huge
del X,y
gc.collect()
print("Memory cleared!")
```

## Preprocessing:

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

cat_pipe = Pipeline([
('encoder', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

preprocessor = ColumnTransformer(
    transformers=[('cat', cat_pipe, cat_cols)],
    remainder='passthrough'
)
```

## Tuned XGB:
XGBoost is being used since it provides inbuilt categorical handling, GPU support etc...

```python
tuned_xgb_params = {'n_estimators': 2382,
                    'learning_rate': 0.036989524346778366,
                    'max_depth': 12,
                    'min_child_weight': 4,
                    'gamma': 2.2246104780594558,
                    'subsample': 0.610932734562846,
                    'colsample_bytree': 0.9417799976499887, 
                    'reg_alpha': 4.766081395540318e-06,
                    'reg_lambda': 1.0193028680619835e-08}

default_params = {
        "random_state": 42,
        "eval_metric": "logloss",
        "tree_method": "hist",   
        "verbosity": 0,
        "device":'cuda',
}
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    **tuned_xgb_params,
    **default_params
)

model_pipeline = Pipeline([
    ('prep', preprocessor),
    ('model',model)
])
```

```python
model_pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import fbeta_score

#finding the best threshold
y_val_perd_probs = model_pipeline.predict_proba(X_val)[:, 1]

best_score = 0
best_thresh = 0

for t in [i/100 for i in range(10, 90)]:
    preds = (y_val_perd_probs > t).astype(int)
    score = fbeta_score(y_val, preds, beta=2) #bets=2 stands for f2_score which is our metric for this competition
    
    if score > best_score:
        best_score = score
        best_thresh = t

print(best_thresh, best_score)
```

```python
del X_train,y_train,X_val,y_val
gc.collect()
print('Memory cleared!')
```

## Making predictions on test data:

```python
test_path = '/kaggle/input/competitions/cyber-physical-anomaly-detection-for-der-systems/test.csv'

test_df = pl.read_csv(test_path)
print('Test set: ', test_df.shape)
display(test_df.head(3))
```

```python
cols_to_remove = unique_count_1_cols + cols_to_drop + ['Id']
X_test = test_df.drop(cols_to_remove)
```

```python
X_test = create_features(X_test)
```

```python
y_test_proba = model_pipeline.predict_proba(X_test)[:,1]

y_preds = (y_test_proba > best_thresh).astype(int)
```

```python
submission = pl.DataFrame({
    "Id": test_df['Id'],
    "Label": y_preds
})

submission.write_csv("submission.csv")
```

## Feature Importance:

```python
model = model_pipeline.named_steps['model']
preprocessor = model_pipeline.named_steps['prep']

feature_names = preprocessor.get_feature_names_out()

importance = model.get_booster().get_score(importance_type='gain')

# Map f-index to real feature names
mapped_features = [
    feature_names[int(f[1:])] for f in importance.keys()
]

importance_df = pl.DataFrame({
    'feature': mapped_features,
    'importance': list(importance.values())
}).sort('importance', descending=True)

print(importance_df.head(70))
```
